```python
# Importing libraries
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Generating random distributions

# Random normal distribution 1 (mean = 5, std = 0.1)
normDist1 = np.random.normal(loc = 5, scale = 0.1, size = 100000)
# Random normal distribution 2 (mean = 5, std = 0.01)
normDist2 = np.random.normal(loc = 5, scale = 0.01, size = 100000)

# Random exponential distribution 1 (lambda = 0.1)
expDist1 = np.random.exponential(scale = 0.1, size = 100000)
# Random exponential distribution 2 (lambda = 0.01)
expDist2 = np.random.exponential(scale = 0.01, size = 100000)
```

```python
# Plotting histograms and scatter plots of the above distributions

# Normal distribution
# Histogram
plt.hist(normDist1, bins = 100)
plt.hist(normDist2, bins = 100)
```

```python
# Scatter plot of two random normal distributions (normDist1 and normDist2)
plt.scatter(normDist1, normDist2)
```

```python
# Adding histogram with color-bar to visualize scatter plot of two normal distributions
h1 = plt.hist2d(normDist1, normDist2)
plt.colorbar(h1[3])
```

```python
# Exponential distribution
# Histogram
plt.hist(expDist1, bins = 100)
plt.hist(expDist2, bins = 100)
```

```python
# Scatter plot of two random exponential distributions (expDist1 and expDist2)
plt.scatter(expDist1, expDist2)
```

```python
# Adding histogram with color-bar to visualize scatter plot of two exponential distributions
h2 = plt.hist2d(expDist1, expDist2)
plt.colorbar(h2[3])
```
